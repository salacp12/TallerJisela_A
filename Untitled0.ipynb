```python
import numpy as np
import analisis_flexiones

cargas = np.array([10, 20, 15])
longitudes = np.array([2, 3, 4])
inercia = 500
modulo_elastico = 200000
distancia_neutra = 0.5

momentos_flectores = analisis_flexiones.calcular_momento_flector(cargas, longitudes)
curvatura = analisis_flexiones.calcular_curvatura(momentos_flectores, inercia, modulo_elastico)
deformacion = analisis_flexiones.calcular_deformacion(curvatura, distancia_neutra)
tensiones = analisis_flexiones.calcular_tension(momentos_flectores, distancia_neutra, inercia)

print("Momentos flectores:", momentos_flectores)
print("Curvatura:", curvatura)
print("Deformación:", deformacion)
print("Tensiones:",tensiones)
```
